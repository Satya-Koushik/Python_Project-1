**EDA Case Study**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('train.csv')
data
```

```python
data.info()
```

```python
data['Stay_In_Current_City_Years'] = data['Stay_In_Current_City_Years'].replace('4+', '4')
```

```python
data
```

```python
data['Stay_In_Current_City_Years'] = data['Stay_In_Current_City_Years'].astype(int)
```

```python
data.info()
```

**Data Cleaning**

```python
#Dealing with null values
data.isnull().sum()
```

```python
# Convert to numeric type (float) before filling NaN values
data['Product_Category_2'] = data['Product_Category_2'].astype(float).fillna(0)
data['Product_Category_3'] = data['Product_Category_3'].astype(float).fillna(0)
```

```python
data
```

```python
data.isnull().sum()
```

```python
#Dealing with duplicate values
data.duplicated().sum()
```

**Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in data.columns:
  if data[col].dtype == 'object':
    data[col] = le.fit_transform(data[col])
```

```python
data
```

Label encoding is done on the object columns. It converts to the numerical values. So, that they are machine readable.

**Visual Analysis**

```python
sns.countplot(data=data, x='Age', hue='Gender')
plt.show()
```

**Inference**:
people of age group 26-35 had purchased more than other age groups. In that male purchased more than female.

**Inference**:

**Inference**:

```python
plt.figure(figsize=(10,15))
ax1 = plt.subplot(3,1,1)
ax2 = plt.subplot(3,1,2)
ax3 = plt.subplot(3,1,3)
sns.countplot(data=data,x='Product_Category_1',hue='Gender',ax=ax1)
sns.countplot(data=data,x='Product_Category_2',hue='Gender',ax=ax2)
sns.countplot(data=data,x='Product_Category_3',hue='Gender',ax=ax3)
plt.show()
```

**Inference**:

In product category 1 many male people are purchasing 1 or 5 or 8 items.

In product category  2 most of the people are not purchasing even 1 item. Few people are purchasing 2 or 8 or 14 or 15 or 16 items.

In product category 3 Most of the people are not even purchasing 1 item.

**Conclusion**:

1. **Age Group**: People aged 26-35 buy more than any other age group. Among them, males make more purchases than females.
   
2. **Occupation**: People in the 4th occupation category tend to buy more, and again, males are the larger group of buyers compared to females.

3. **Marital Status**: Married males make more purchases than unmarried males.

4. **Budget Preference**: Most people prefer to spend between 5,000 and 10,000 on purchases.

5. **City Preference**: Males in city category 1 tend to make more purchases than those in city categories 0 and 2.

6. **City Duration**: Males who have lived in a city for 1 year tend to purchase more than those who have lived there for other durations.

7. **Product Category 1**: In this category, most males buy 1, 5, or 8 items.

8. **Product Category 2**: Most people in this category don’t buy anything, but a few purchase 2, 8, 14, 15, or 16 items.

9. **Product Category 3**: Most people don’t buy anything in this category.

In summary, males, especially those aged 26-35 and married, tend to make more purchases, and people usually prefer a budget of 5,000-10,000. The purchasing patterns vary across product categories and cities, with city 1 and the 4th occupation category seeing the most purchases.
